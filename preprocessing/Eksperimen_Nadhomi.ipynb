```python
df_feature = ['RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe']

# Proses imputasi
imputer = SimpleImputer(strategy='mean')
df[df_feature] = imputer.fit_transform(df[df_feature])
```

```python
df
```

```python
# Normalisasi
scaler = MinMaxScaler()
df[df_feature] = scaler.fit_transform(df[df_feature])
```

```python
X = df.drop('Type', axis=1)
y = df['Type']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```
